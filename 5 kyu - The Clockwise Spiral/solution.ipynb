Do you know how to make a spiral? Let's test it!
Classic definition: A spiral is a curve which emanates from a central point, getting progressively farther away as it revolves around the point.

Your objective is to complete a function createSpiral(N) that receives an integer N and returns an NxN two-dimensional array with numbers 1 through NxN represented as a clockwise spiral.

Return an empty array if N < 1 or N is not int / number

Examples:

N = 3 Output: [[1,2,3],[8,9,4],[7,6,5]]

1    2    3    
8    9    4    
7    6    5    
N = 4 Output: [[1,2,3,4],[12,13,14,5],[11,16,15,6],[10,9,8,7]]

1   2   3   4
12  13  14  5
11  16  15  6
10  9   8   7
N = 5 Output: [[1,2,3,4,5],[16,17,18,19,6],[15,24,25,20,7],[14,23,22,21,8],[13,12,11,10,9]]

1   2   3   4   5    
16  17  18  19  6    
15  24  25  20  7    
14  23  22  21  8    
13  12  11  10  9

```python
import numpy as np
def create_spiral(n):
    if not isinstance(n, int) or n < 1:
        return []
    arr = np.ones((n,n), dtype=int)
    r_sides = [0,n]
    s_sides = [0,n]
    pointer = [0,0]
    direction = [0,1]
    cnt = 1
    while cnt < n**2:
        while r_sides[0] <= pointer[1] < r_sides[1] and s_sides[0] <= pointer[0] < s_sides[1]:
            arr[pointer[0], pointer[1]] = cnt
            cnt += 1
            pointer[0] += direction[0]
            pointer[1] += direction[1]
        cnt -= 1
        if direction[1] == 1:
            pointer[1] -= 1
            direction[1] = 0
            direction[0] = 1
        elif direction[0] == 1:
            pointer[0] -= 1
            direction[1] = -1
            direction[0] = 0
        elif direction[1] == -1:
            pointer[1] += 1
            s_sides[0] += 1
            direction[1] = 0
            direction[0] = -1
        elif direction[0] == -1:
            direction[1] = 1
            direction[0] = 0
            pointer[0] += 1
            pointer[1] += 1
            cnt += 1
            arr[pointer[0], pointer[1]] = cnt
            s_sides[1] -= 1
            r_sides[0] += 1
            r_sides[1] -= 1
    return arr.tolist()
```
